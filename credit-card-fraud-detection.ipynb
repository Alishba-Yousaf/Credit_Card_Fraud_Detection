#  **Statystical Analysis**

```python
df = pd.read_csv('/kaggle/input/datasets/organizations/mlg-ulb/creditcardfraud/creditcard.csv')

df.head()
```

**Observation**

The dataset contains 284,807 transactions with 31 features.
There are no missing values in the dataset.
Most features (V1–V28) are anonymized numerical variables obtained using PCA transformation.
The target variable 'Class' represents whether a transaction is fraudulent (1) or normal (0).

# Visual Analysis

## Class Distribution (Fraud vs Normal Transactions)

```python
import matplotlib.pyplot as plt

class_counts = df['Class'].value_counts()

plt.figure(figsize=(6,4))
bars = plt.bar(['Normal','Fraud'], class_counts)

plt.title("Fraud vs Normal Transactions")
plt.xlabel("Transaction Type")
plt.ylabel("Number of Transactions")

for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 500, int(yval), 
             ha='center', va='bottom')

plt.show()
```

**Observation**

The dataset is highly imbalanced. Normal transactions are significantly higher than fraudulent transactions. Only a very small portion of transactions are fraud, which makes fraud detection a challenging machine learning problem.

## Transaction Amount Distribution

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df['Amount'], bins=50)

plt.title("Transaction Amount Distribution")
plt.xlabel("Transaction Amount")
plt.ylabel("Frequency")

plt.show()
```

**Observation**

Most transactions involve small amounts, while very large transactions occur rarely.
The distribution is highly right-skewed, indicating that the majority of credit card
transactions are of lower value.

## Transaction Time Distribution

```python
plt.figure(figsize=(8,5))

plt.hist(df['Time'], bins=50)

plt.title("Transaction Time Distribution")
plt.xlabel("Time")
plt.ylabel("Frequency")

plt.show()
```

**Observation**

Transactions are distributed across the entire time period in the dataset.
This indicates continuous transaction activity over time.

## Fraud vs Transaction Amount

```python
plt.figure(figsize=(8,5))

plt.scatter(df['Amount'], df['Class'], alpha=0.5)

plt.title("Fraud vs Transaction Amount")
plt.xlabel("Transaction Amount")
plt.ylabel("Class (0 = Normal, 1 = Fraud)")

plt.show()
```

**Observation**

Fraudulent transactions occur across different transaction amounts.
There is no single amount range where fraud occurs exclusively.

## Feature Correlation Heatmap

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(14,10))

sns.heatmap(df.corr(), cmap='coolwarm')

plt.title("Feature Correlation Heatmap")

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```

**Observation**

The heatmap shows the correlation between different features in the dataset.
Most features have low correlation with each other because they are anonymized PCA transformed variables.

# Data Preprocessing

## Feature Scaling (Amount)

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

df['Amount_scaled'] = scaler.fit_transform(df[['Amount']])

df.head()
```

**Reason**

The Amount feature has a wide range of values. 
Scaling is applied to normalize the values so that machine learning models can perform better.

## Removing Original Amount Feature

```python
df = df.drop(['Amount'], axis=1)

df.head()
```

**Reason**

The original Amount feature is removed after scaling to avoid redundancy.
The scaled feature (Amount_scaled) will be used for further analysis and model training.

## Class Imbalance Check

```python
df['Class'].value_counts()
```

**Reason**

This step confirms that the dataset is highly imbalanced.
The number of normal transactions is much higher than fraudulent transactions,
which is an important factor to consider during model training.

## Save Preprocessed Dataset

```python
df.to_csv("preprocessed_creditcard.csv", index=False)
```

**Reason**

The preprocessed dataset is saved in CSV format so that it can be directly used
in the next phases of the project without repeating visualization and preprocessing steps.
